```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv

# Set up options
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Run Chrome in headless mode
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Set paths for Chromedriver and Chromium
chromedriver_path = r'C:\Program Files\Chromedriver\chromedriver.exe'
chromium_path = r'C:\Program Files\Chromium\chrome.exe'

# Initialize the WebDriver with the correct service
service = ChromeService(executable_path=chromedriver_path, log_path='chromedriver.log')
driver = webdriver.Chrome(service=service, options=options, executable_path=chromium_path)

# URL of the website to scrape
url = 'https://www.marinetraffic.com/en/maritime-news#google_vignette'
driver.get(url)

# Add a wait for elements to load
wait = WebDriverWait(driver, 10)  # Wait for 10 seconds for elements to load
try:
    news_items = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.MuiBox-root.css-1ivgyjj')))  # Adjust as necessary
except Exception as e:
    print(f"Error finding news items: {e}")
    driver.quit()
    exit()

# Open a CSV file to write data
with open('maritime_news.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Date', 'Title', 'Content'])

    # Limit the number of articles scraped (adjust limit as needed)
    for item in news_items[:10]:  # Scrape only the first 10 articles
        try:
            # Extract date (replace with appropriate selector)
            date_element = item.find_element(By.CSS_SELECTOR, '.MuiTypography-root.MuiTypography-body2.css-igvigi')  # Adjust as necessary
            date = date_element.text.split("on ")[1]  # Extract date after 'on'

            # Extract title (replace with appropriate selector)
            title_element = item.find_element(By.CSS_SELECTOR, '.MuiTypography-root.MuiTypography-h2.css-oods9f')  # Adjust as necessary
            title = title_element.text

            # Extract content (replace with appropriate selector for main content paragraph)
            content_elements = item.find_elements(By.CSS_SELECTOR, '.MuiBox-root.css-1ivgyjj div p:not(:first-child)')  # Select all paragraphs except the first one
            content_paragraphs = [p.text.strip() for p in content_elements]  # Extract and clean each paragraph
            content = "\n".join(content_paragraphs)  # Join paragraphs back with newline

            writer.writerow([date, title, content])
        except Exception as e:
            print(f"Error extracting data from item: {e}")

# Close the driver
driver.quit()

print("Web scraping completed and data saved to maritime_news.csv")
```
